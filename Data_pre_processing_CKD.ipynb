```python
df.EG_White = df.EG_White.replace(2,0)
df.EG_BAME = df.EG_BAME.replace(2,0)
df.EG_Other = df.EG_Other.replace(2,0)
df.Male = df.Male.replace(2,0)
#### Drop female coulmn and rename male column as gneder with 1 = male, 0=female
df = df.drop(['Female'], axis=1)
df = df.rename({'Male': 'Gender'}, axis=1)

df.Comorbidity = df.Comorbidity.replace(2,0)
df.Bleeding_Risk = df.Bleeding_Risk.replace(2,0)
df.VTE_Risk = df.VTE_Risk.replace(2,0)
df.Apixa_2_5mg = df.Apixa_2_5mg.replace(2,0)
df.Apixa_5mg = df.Apixa_5mg.replace(2,0)
df.Apixa_10mg = df.Apixa_10mg.replace(2,0)
df.Dabi_75mg = df.Dabi_75mg.replace(2,0)
df.Dabi_110mg = df.Dabi_110mg.replace(2,0)
df.Dabi_150mg = df.Dabi_150mg.replace(2,0)
df.Edoxa_30mg = df.Edoxa_30mg.replace(2,0)
df.Edoxa_60 = df.Edoxa_60.replace(2,0)
df.Riva_10mg = df.Riva_10mg.replace(2,0)
df.Riva_15mg = df.Riva_15mg.replace(2,0)
df.Riva_20mg = df.Riva_20mg.replace(2,0)
df.Normal_Renal_Function = df.Normal_Renal_Function.replace(2,0)
df.Stage3a_CKD = df.Stage3a_CKD.replace(2,0)
df.Stage3b_CKD = df.Stage3b_CKD.replace(2,0)
df.Stage4_CKD = df.Stage4_CKD.replace(2,0)
df.Stage5_CKD = df.Stage5_CKD.replace(2,0)
```
